**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
# отфильтруем данные по признаку salary
over_50k = data[data['salary'] == '>50K'] 
under_50k = data[data['salary'] == '<=50K']

avg_over_50k = over_50k['age'].mean()
avg_under_50k = under_50k['age'].mean()

std_over_50k = over_50k['age'].std()
std_under_50k = under_50k['age'].std()
print(f'Среднее значение возраста тех кто получает больше 50К: {avg_over_50k:.2f}')
print(f'Среднее значение возраста тех кто получает меньше 50К: {avg_under_50k:.2f}')
print(f'Среднеквадротичное отклонение тех кто получает больше 50К: {std_over_50k:.2f}')
print(f'Среднеквадротичное отклонение тех кто получает меньше 50К: {std_under_50k:.2f}')
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# создадим столбец указывающий женат ли мужчина
males['is_married'] = males['marital-status'].apply(lambda x: x.startswith('Married'))

# сгрупируем всё по этому столбцу и найдём всех с зп >50К
married_males_with_over_50k = males[males['salary'] == '>50K'].groupby('is_married').size()

# расчет доли для каждой группы
proportions = married_males_with_over_50k/males.groupby('is_married').size()

print(proportions)
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
def age_grouping(age):
    if 16 <= age < 35:
        return 'young'
    elif 35 <= age < 70:
        return 'adult'
    elif 70 <= age <= 100:
        return 'retiree'
    else:
        return 'other'
    
data['AgeGroup'] = data['age'].apply(age_grouping)
data.head()
```

```python
# также можно создать словарь с возрастными группами если их будет больше 3
age_groups = {
    'young': (16, 35),
    'adult': (36, 70),
    'retiree': (71, 100)
}
# Новый вид функции
def age_grouping_by_dict(age):
    for group, (min_age, max_age) in age_groups.items():
        if min_age <= age <= max_age:
            return group

data['AgeGroup'] = data['age'].apply(age_grouping_by_dict)
data.head()
```

```python
# использование втроенного метода pandas - cut() 
bins = [15, 35, 70, 100] # возрасты
labels = ['young', 'adult', 'retiree'] # группы

data['AgeGroup'] = pd.cut(data['age'], bins=bins, labels=labels)
data.head()
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
over_50k = data[data['salary'] == '>50K']
age_groupe_count_over_50k = over_50k['AgeGroup'].value_counts()
print(age_groupe_count_over_50k)

group_with_greatest_count = age_groupe_count_over_50k.idxmax()
print('\nГруппа с наибольшим количеством зарабатывающих >50K:', group_with_greatest_count)
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
occupation_groups = data.groupby('occupation')

def filter_func(group):
    return group['age'].mean() <= 40 and group['hours-per-week'].min() > 5

filtered = occupation_groups.filter(filter_func)

filter_counts = filtered['occupation'].value_counts()
filter_counts
```
